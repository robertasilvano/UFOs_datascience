```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('pre_processado_completo_mapa.csv', header=0,error_bad_lines=False, warn_bad_lines=False, low_memory=False)
df.head()
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

```python
lat_lon = df[['latitude', 'longitude']]
lat_lon = lat_lon.to_records(index=False)
lat_lon = list(lat_lon)
lat_lon
```

```python
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
```

```python
locations = [reverse(loc) for loc in lat_lon if loc else 0]
```

```python
countrys = []
states = []
for row in df.itertuples():
    lat = str(row[8])
    lon = str(row[9])
    
    location = geolocator.reverse(lat+","+lon)
    address = location.raw['address']
    country = address.get('country', '')
    state = address.get('state', '')
    countrys.append(country)
    states.append(state)
```

```python
df['Country_new'] = countrys
df['States_new'] = states
```
